FREQUENCY TABLE - ONE-WAY TABLE

```python
pd.crosstab(index=car_backup['FuelType'],columns='Count',dropna=True)
```

FREQUENCY TABLE - TWO-WAY TABLE

```python
pd.crosstab(index=car_backup['Automatic'], columns=car_backup['FuelType'],dropna=True)
```

jOINT PROBABILITY USING TWO-WAY TABLE

```python
pd.crosstab(index=car_backup['FuelType'],columns=car_backup['Automatic'],normalize=True,dropna=True)
```

Marginal Probability using Two-way table# 

# Marginal Probability using Two-way table

```python
pd.crosstab(index=car_backup['FuelType'],columns=car_backup['Automatic'], margins=True, normalize=True, dropna=True)
```

# Conditional Probability using Two-way table
Given the type of gearbox, what is the probability of different fuel types?

```python
pd.crosstab(index=car_backup['Automatic'], columns=car_backup['FuelType'],normalize=True, margins=True, dropna=True)
```
